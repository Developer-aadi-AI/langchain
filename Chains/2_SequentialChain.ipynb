```python
from langchain_ollama import ChatOllama
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
```

```python
model = "llama3.2"
base_url = "http://localhost:11434"

llm = ChatOllama(
    model = model,
    base_url = base_url,
    temperature=0.7
)

prompt1 = PromptTemplate(template="write a detailed report about the performance of {player}",
                        input_variables=['player'],
                        )

prompt2 = PromptTemplate(template="Generste five pointer summary {text}",
                        input_variables=['text']
                        )

parser = StrOutputParser()

chain = prompt1 | llm | parser | prompt2 | llm | parser

result = chain.invoke({"player":"MS Dhoni"})

print(result)
```
