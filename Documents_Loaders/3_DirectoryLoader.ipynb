```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader
```

```python
loader = DirectoryLoader(
    path = "Books", # to give path of directory
    glob="*.pdf", # to load all pdf files
    loader_cls=PyPDFLoader
    )

docs = loader.load()

print(len(docs))   ## This will be equal to no. of pages combined from the pdf
```
