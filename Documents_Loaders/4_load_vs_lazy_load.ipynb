```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader
```

```python
loader = DirectoryLoader(
    path="Books",
    glob="*.pdf",
    loader_cls=PyPDFLoader
    )

docs = loader.load()

for doc in docs:
    print(doc.metadata)  #This will take a lot of time to run
```

```python
docs = loader.lazy_load()

for doc in docs:
    print(doc.metadata)  #This takes less time and memory as single document is loaded at once. 
```
