## JsonOutputParser

Here in this code we will run how to use JsonOutputParser

### Import Libraries

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
```

### Set model

```python
model = "llama3.2"
local_url = "http://localhost:11434"

llm = ChatOllama(
    model = model,
    local_url=local_url,
    temperature=0.3)
```

### Call Parser

```python
parser = JsonOutputParser()
```

### build prompt template

```python
template = PromptTemplate(template="Give me the name age and city of a fictional person,\n {format_instruction}",
                        input_variables=[],
                        partial_variables={'format_instruction': parser.get_format_instructions()})
prompt=template.format()
print(prompt)

```

### Invoke Model :- Without Chains

```python
result = llm.invoke(prompt)

final_result = parser.parse(result.content)

print(final_result)
print(type(final_result))  ## It will resturn dict, because python treats json as dict 
```

### Invoke Models :- with Chains

```python
template = PromptTemplate(template="Give me the name age and city of a fictional person,\n {format_instruction}",
                        input_variables=[],
                        partial_variables={'format_instruction': parser.get_format_instructions()})

chain = template | llm | parser
# result = chain.invoke() # this gives an error of reqired one positional argument 'input' so pass an empty dictionary
result = chain.invoke({})
print(result)
```

### Problem with JsonOutputParser

We can not enforce any schema/structure that should be followed in our output 


### let's take an example

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
```

```python
model = "llama3.2"
local_url = "http://localhost:11434"

llm = ChatOllama(
    model = model,
    local_url=local_url,
    temperature=0.3)

parser = JsonOutputParser()


template = PromptTemplate(template="Give me a five bullet point summary about {topic} and every point should be less than 10 words,\n {format_instruction}",
                        input_variables=['topic'],
                        partial_variables={'format_instruction': parser.get_format_instructions()})

chain = template | llm | parser
# result = chain.invoke() # this gives an error of reqired one positional argument 'input' so pass an empty dictionary
result = chain.invoke({'topic':'langchain'})
print(result)

## It did not game me the summary in bullet points and i can not get it like in a structured way.

```
