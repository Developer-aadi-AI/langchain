## StrOutputParser

In this file I am going to show you the usecase of StrOutputParser, that how we can get text input from StrOutputParsers.

first lets start with our basic way i.e, result.content

### Import Necessary Libraries

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
```

### Model Setting

```python
model = "llama3.2"
local_url = "http://localhost:11434"

llm = ChatOllama(
    model=model, 
    local_url=local_url,
    temperature=0.7)

```

### Detailed Prompts

```python
# first prompt for report
template1 = PromptTemplate(
    template="write a detailed report on {topic}",
    input_variables=['topic']
)

# second Prompt for summary
template2 = PromptTemplate(
    template="Write a five Bullet Point summary on the following text and each bullet point length should be less than 15 words./n  {text}",
    input_variables=['text']

)
```

### Invoking 1st prompt and model

```python
prompt1 = template1.invoke({'topic' : 'Milkyway Galaxy' })

result1 = llm.invoke(prompt1)
```

```python
prompt2 = template2.invoke({'text' : result1.content})

result2 = llm.invoke(prompt2)
```

```python
print(result2.content)
```

### Now let's work with StrOutputParser

```python
from langchain_ollama import ChatOllama
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
```

```python
model = "llama3.2"
local_url = "http://localhost:11434"

llm = ChatOllama(
    model=model,
    local_url=local_url,
    temperature=0.8)

```

```python
# first prompt for report
template1 = PromptTemplate(
    template="write a detailed report on {topic}",
    input_variables=['topic']
)

# second Prompt for summary
template2 = PromptTemplate(
    template="Write a five Bullet Point summary on the following text and each bullet point length should be less than 15 words./n  {text}",
    input_variables=['text']

)
```

### Building Parser

```python
parser= StrOutputParser()
```

### Building Chain

```python
chain = template1 | llm | parser | template2 | llm | parser
```

### Invoking Chain

```python
result = chain.invoke({'topic' : 'Milkyway'})
```

### Printing Result

```python
print(result)
```
