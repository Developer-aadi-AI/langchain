## ToolKit

```python
from langchain_core.tools import tool
```

```python
@tool
def add(a: int, b: int) -> int:
    """Add two Numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two Numbers"""
    return a * b
```

```python

class MathToolKit:
    def get_tools(self):
        return[add, multiply]
    
toolkit = MathToolKit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
    print(tool.invoke({"a":12, "b":13}))
```
