### Method 1: Using tool library 

```python
from langchain.tools import tool

# step 1 - create a function
def multiply(a,b):
    """Multiply two numbers"""
    return a * b
```

```python
# Step 2: Add type hints to your function
def multiply(a: int,b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
# step 3: add tool decorator
@tool
def multiply(a: int,b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
# Step 4: It is a tool, hence we can use invoke with it.

result = multiply.invoke({"a": 5, "b": 9})
print(result)
```

```python
# step 5: We can also use some tool functions
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
# How is a tool sended to an llm: it is sended in json format that is given below
print(multiply.args_schema.model_json_schema())
```

### Method 2: Using StructuredTool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field 
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required = True, description="The first number to multiply")
    b: int = Field(required=True, description="The second number to multiply")
    
```

```python
def multiply_func(a: int, b: int) -> int:
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="Multiply Two Numbers",
    args_schema=MultiplyInput
)
```

```python

res = multiply_tool.invoke({"a":5,"b":12})
print(res)

print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

### Method 3: Using BaseTool Class

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to multiply")
    b: int = Field(required=True, description="The Second Number to multiply")
```

```python
class MultiplyTool(BaseTool):
    name: str = "Multiplier"
    description: str = "Multiplies two numbers"
    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int,b: int) -> int:
        return a * b
```

```python
multi = MultiplyTool()

res = multi.invoke({"a": 4, "b": 8})
print(res)

print(multi.name)
print(multi.description)
print(multi.args)
```
