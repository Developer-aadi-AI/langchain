### These are the most important steps to follow while dealing with tools

```python
from langchain_core.tools import tool
from langchain_ollama import ChatOllama
import requests
from langchain_core.messages import HumanMessage
```

### Tool Creation

```python
@tool
def multiply(a: int, b: int) ->int :
    """This function takes two integer arguments and will return product as an integer"""
    return a * b
```

```python
llm = ChatOllama(
    base_url="http//:localhost:11434",
    model = "llama3.2",
    temperature=0.3
)
```

### Tool Binding

```python
llm_with_tools = llm.bind_tools([multiply])
```

### Tool Calling

```python
llm_with_tools.invoke("What is 2 * 4")
```

### When we invoke our llm, the llm will not call a tool, instead it gives us an object wich contains the required data for tool as we can see in ex below:  
tool_calls=[{'name': 'multiply', 'args': {'a': 2, 'b': 4}, 'id': 'a9e2f08c-e437-4f05-8da6-1b1c59114bb1', 'type': 'tool_call'}], invalid_tool_calls=[], usage_metadata={'input_tokens': 162, 'output_tokens': 22, 'total_tokens': 184})

```python
llm_with_tools.invoke("what is the product of 1000 and 3").tool_calls
```

### Tool Execution

```python
result = llm_with_tools.invoke("what is the product of 1000 and 3").tool_calls
```

```python
print(multiply.invoke(result[0]["args"]))
```

### Professional way of tool calling with history

```python
query = HumanMessage("Can you multiply 5 and 25")
```

```python
messages = [query]
```

```python
result = llm_with_tools.invoke(messages)
result
```

```python
messages.append(result)
```

```python
tool_result = multiply.invoke(result.tool_calls[0])
tool_result
```

```python
messages.append(tool_result)
```

```python
print(llm.invoke(messages).content)
```
