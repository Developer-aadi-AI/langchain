```python
from langchain_groq import ChatGroq
from langchain_core.tools import tool
import requests
import json
from langchain_community.tools import DuckDuckGoSearchRun
from dotenv import load_dotenv
from langchain.agents import create_agent
```

```python
load_dotenv()
search_tool = DuckDuckGoSearchRun()
```

```python
@tool
def weather(city: str) -> str:
    """This returns the weather of a city, upon given a city name"""
    url = f"https://api.weatherstack.com/current?access_key=3fc90c0ddf68213011493b37ff62fb37&query={city}"

    response = requests.get(url)
    return json.dumps(response.json())
```

```python

agent = create_agent(
    model="groq:llama-3.3-70b-versatile",
    tools=[search_tool, weather],
    system_prompt=(
        "You are a helpful assistant.\n"
        "Always use tools when needed.\n"
        "When calling tools, strictly follow JSON format.\n"
        "Do not output anything except valid tool calls when required."
    )
)
```

```python
response = agent.invoke({
    "messages": [
        {"role": "user", "content": "Find the capital of India, then find its current weather condition"}
    ]
})
```
