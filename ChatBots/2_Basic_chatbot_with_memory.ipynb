```python
from langchain_ollama import ChatOllama
```

```python
base_url = "http://localhost:11434"
model = "llama3.2"
llm = ChatOllama(
    base_url=base_url,
    model = model,
    temperature = 0.8,
    num_predict=300
)

memory = []

while True:
    prompt  = input("User : ")
    if prompt.lower() == "exit":
        break
    memory.append(prompt)
    result = llm.invoke(memory)
    print("AI : ",result.content)
    memory.append(result.content)
```

### As we can see if we use a list to store the output, the model get's confused and hallucinate between User message and AI message,   
### to overcome this we can use dictionary, but we will have to manually update the messages in dictionary as well, 

### well, Langchain has solved this problem 
- we can use messages  
    - SystemMessage
    - HumanMessage
    - AIMessage
