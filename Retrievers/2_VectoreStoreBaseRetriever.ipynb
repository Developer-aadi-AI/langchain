```python
from langchain_community.vectorstores import Chroma
from langchain_ollama import OllamaEmbeddings
from langchain_core.documents import Document
```

```python
# Step 1: Your source documents
documents = [
    Document(page_content="LangChain helps developers build LLM applications easily."),
    Document(page_content="Chroma is a vector database optimized for LLM-based search."),
    Document(page_content="Embeddings convert text into high-dimensional vectors."),
    Document(page_content="OpenAI provides powerful embedding models."),
]
```

```python
# setting embedding model
embedding_model = OllamaEmbeddings(model = 'nomic-embed-text')

# Create Chroma Vector Store in Memory 
vectorstore = Chroma.from_documents(
    documents=documents,
    embedding = embedding_model,
    collection_name="my_collection"
)
```

```python
retriever = vectorstore.as_retriever(search_kwargs = {"k":2})
```

```python
query = "What is Chroma used for?"
results = retriever.invoke(query)
```

```python
print(results)
```

```python
for i, doc in enumerate(results):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content)
```

```python
results = vectorstore.similarity_search(query, k=2)  # The above result can be generated by similarity search as well but we can not use different techniques with similarity search but we can with VectorStoreBaseRetriever
```

```python
for i, doc in enumerate(results):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content)
```
